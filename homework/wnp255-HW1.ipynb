```python
import random
import numpy as np
import matplotlib.pyplot as plt
def approximatePi(numSamples : int):
    counter = 0
    
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        
        if ((x*x) + (y*y) < 1):
            counter +=1
            
    return 4.0 * counter / numSamples
```

```python
#Our level of accuracy
x = np.arange(10,10000,20)
#Error relative to the actual value
y = [(abs(np.pi - approximatePi(num))) for num in x]
```

```python
#plotting our error
plt.plot(x,y)
plt.xlabel("Number of Samples")
plt.ylabel("Absolute error |pi - pi^|")
plt.title("Absolute error as function of #samples for approximating pi")
```

```python
#Our level of accuracy
x = np.arange(10,10001,20)
#Error relative to the actual value
y = [(abs(np.pi - approximatePi(num))/np.pi) for num in x]
```

```python
#plotting our error
plt.plot(x,y)
plt.xlabel("Number of Samples")
plt.ylabel("Relative error |pi - pi^|/pi")
plt.title("Relative error as function of #samples for approximating pi")
```
