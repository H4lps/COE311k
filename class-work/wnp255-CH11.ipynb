```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.interpolate import CubicSpline
```

```python
def mse(y,y_pred):
    mse = np.mean((y - y_pred)**2)
    return mse
```

```python
def r_squared(y,y_pred):
    ss_total = np.sum((y - np.mean(y))**2)
    ss_res = np.sum((y - y_pred)**2)
    r_squared = 1 - (ss_res / ss_total)
    return r_squared
```

```python

# Define known data points

x = np.array([0.1, 1, 2, 3, 4, 5])

y = np.array([0.1, 1, 0.1, 1, 0.1, 1])

# Create a cubic spline interpolation of the data

cs = CubicSpline(x, y, bc_type='natural')# Generate new x values for a smooth plot

x_interp = np.linspace(0.1, 5, 100)

y_interp = cs(x)

#Polyfit
poly_coeffs = np.polyfit(x,y,deg=3)
y_pred = np.polyval(poly_coeffs[::-1],x)

#compare
mse_cs = mse(y,y_interp)
mse_p = mse(y,y_pred)

r_squared_cs = r_squared(y,y_interp)
r_squared_p = r_squared(y,y_pred)

if r_squared_cs > r_squared_p:
    print("The cubic spline is better in terms of the r squared quantity")
else:
    print("The polynomial fit is better in terms of the r squared quantity")
if mse_cs > mse_p:
    print("The cubic spline is better in terms of the r squared quantity")
else:
    print("The polynomial fit is better in terms of the r squared quantity")
```
