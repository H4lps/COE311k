```python
import random
import numpy as np
import matplotlib.pyplot as plt
def approximatePi(numSamples : int):
    counter = 0
    
    for i in range(numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        
        if ((x*x) + (y*y) < 1):
            counter +=1
            
    return 4.0 * counter / numSamples
```

```python
abs_errs = []
rel_errs = []

for i in range(100):
    aprox_pi = approximatePi(1000)
    abs_err = abs(np.pi - aprox_pi)
    rel_err = abs_err/np.pi
    
    abs_errs.append(abs_err)
    rel_errs.append(rel_err)
```

```python
plt.plot(np.arange(0,100),abs_errs)
plt.xlabel("Run #")
plt.ylabel("|pi - aprox_pi|")
plt.title("Absolute error over 100 MC runs")
```

```python
plt.plot(np.arange(0,100),rel_errs)
plt.xlabel("Run #")
plt.ylabel("|pi - aprox_pi|/pi")
plt.title("Relative error over 100 MC runs")
```
