```python
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
#setting up subplot configuration
fig, axes = plt.subplots(nrows = 2 ,ncols = 2)
```

```python
#Plot 1
ax1 = axes[0,0]
x = np.linspace(0,2*np.pi,50)
plt.plot(x, np.sin(x), 'r-o',label="sinx")
plt.plot(x, np.cos(x), 'g--',label="cosx")
plt.legend()
plt.title("sin(x) plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
#Plot 2
ax2 = axes[0,1]
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
plt.scatter(x, y, size, color)
plt.colorbar()
plt.show()
```

```python
#Plot 3
ax3 = axes[1,0]
x = np.random.rand(1000)
y = np.random.rand(1000)
plt.hist(x, 50)
plt.title("Histogram of random distribution")
plt.ylabel("Counts")
plt.xlabel("bins")
plt.show()
```

```python
#plot 4
def f1(x,y):
    return (x*x+y*y)
x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(45, 0)
```
